```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://en.m.wikipedia.org/wiki/Stealth_aircraft"
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
table=soup.find_all('table')[0]
```

```python
titles=table.find_all('th')
titles
```

```python
heads=[i.text.strip() for i in titles]
heads
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=heads)
df
```

```python
contents=table.find_all('tr')[1:]

print(len(contents))
for row in contents:
  data=row.find_all('td')
  infos=[i.text.strip() for i in data]
  length=len(df)
  print(length)
  print(infos)
  if len(infos)==len(df.columns):
    df.loc[length]=infos
df
```

```python
df.to_csv('stealth_aircraft.csv',index=False)
```
